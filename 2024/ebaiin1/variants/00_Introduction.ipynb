# Introduction

## I. Définition

### 1. Qu'est ce que c'est une variation génomique ?

Une variation génomique est un changement d'une ou plusieurs bases nucléotides, dans une séquence d'ADN particulière en comparaison avec une séquence d'ADN (un génome) de référence. Les variations génomiques se distinguent en deux catégories : Polymorphismes et mutations.

### 2. Différents types de variations 

- **SNV** : **S**ingle **N**ucleotide **V**ariant
- **INDEL** : **IN**sertion ou **DEL**etion d'une ou plusieurs bases
- **MNV** : **M**ulti-**N**ucleotide **V**ariant, soit plusieurs SNVs et/ou INDELs dans un bloc
- **SV** : **S**tructural **V**ariant, réarragement génomique affectant > 50bp
- **CNV** : **C**opy **N**umber **V**ariation

## II. SNV ≠ SNP


- **SNV** (**S**ingle **N**ucleotide **V**ariant) : Toute altération nucléotidique sans implication de fréquence populationnelle
- **SNP** (**S**ingle **N**ucleotide **P**olymorphism) : Implique qu'un variant est partagée dans la population (> 1%)

/!\ L'amalgame entre SNPs et SNVs est très souvent fait
