# Bonus : Gameshell

Pour pratiquer toutes les commandes bash présentées (et quelques autres) sous la forme d'un jeu "Donjon et Dragon" (mais sans dragon ...), vous pouvez lancer Gameshell https://github.com/phyver/GameShell :

- ouvrir un **terminal**   
- aller dans votre espace projet : <code>cd #espace_projet</code>  
- créer un répertoire d'accueil du jeu : <code>mkdir -p Gameshell</code>  
- et s'y placer : <code>cd Gameshell</code>   
- télécharger le jeu :   
<code>wget https://github.com/phyver/GameShell/releases/download/latest/gameshell.sh</code>  
- lancer le jeu : <code>bash gameshell.sh</code>

Quand on quitte le jeu, la session est sauvegardée et donc pour retrouver sa session, il suffit de relancer le jeu (<code>bash gamesell.sh</code>)

Pour le jouer en anglais (en) ou en italien (it) : <code>bash -L en gameshell.sh</code>
