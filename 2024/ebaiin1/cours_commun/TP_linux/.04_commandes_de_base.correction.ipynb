## Créer de nouveaux dossiers

Maintenant que nous savons où écrire, nous allons essayer de créer un nouveau dossier.

La commande est simple : <code>mkdir nom_du_dossier</code> (_make directory_)

Vérifiez que vous êtes dans votre dossier projet, sous-dossiers <code>cours_commun/TP_linux</code>

```bash
# Afficher le chemin absolu où l'on se trouve
pwd
```

Créez le dossier <code>chip-seq</code>

```bash
# Création du dossier chip-seq
mkdir chip-seq
```

Vérifiez le contenu de votre dossier pour voir si le nouveau dossier chip-seq est bien apparu.
##### Tips : Si vous ajoutez l'option <code>-p</code> à votre commande, les dossiers apparaissent avec un <code>/</code> à la fin, les rendant plus facilement identifiable !

```bash
# Contenu du dossier
ls -l
```

S'il est possible de créer un dossier vide, il est possible de le supprimer ! 

La commande est presque la même : <code>rmdir nom_du_dossier</code> (_remove directory_)

Essayons de supprimer le dossier <code>chip-seq</code> pour créer un nouveau dossier <code>rna-seq</code> (vous devez obtenir deux lignes de commande)

```bash
# Suppression du dossier vide
rmdir chip-seq
# Création du nouveau dossier
mkdir rna-seq
```

Afin d'organiser votre dossier vous vous déplacez dedans 

```bash
# Déplacement dans le nouveau dossier créé
cd rna-seq
```

Vous créez un dossier <code>ref</code> et un <code>fasta</code> (en une seule ligne)

```bash
# Création de deux dossiers d'un coup
mkdir ref fasta 
```

Vérifiez que les dossiers se sont bien créés

```bash
# Liste du contenu
ls -l
```

Parfait, maintenant remontons d'un cran dans l'arborescence !

```bash
# Déplacement un cran en arrière
cd ..
```

Sans changer d'emplacement, vous pouvez créer un dossier et l'ensemble de ses sous-dossiers. Pour cela, on utilise l'option <code>-p</code> et on écrit le chemin que l'on souhaite créer.

Essayons donc de créer le **dossier** <code>output</code> et à l'intérieur le **sous-dossier** <code>bam</code>, sans utiliser la commande <code>cd</code> !

```bash
# Création d'une succession de dossiers imbriqués
mkdir -p rna-seq/output/bam
```

Quelle commande utiliser pour vérifier notre arborescence complète ?

```bash
# Afficher l'arborescence
tree
```

## Copier, coller et supprimer des fichiers ou dossiers

##### ATTENTION : La ligne de commande permet beaucoup de chose et est beaucoup moins protégée que le clique-bouton de Windows. 

##### Le shell ne vous demandera pas 20 fois si vous êtes sûr.e.s de vouloir écraser un fichier, il le fera.

##### Ainsi, la plus grande prudence est de mise !

### Copier/coller un fichier

<code>cp</code> est l'équivalent du copier-coller.

Elle prend obligatoirement deux arguments : <code>cp fichier_source destination</code>

La destination peut être le nouveau nom de notre fichier si nous ne souhaitons pas garder son nom d'origine. Cela permet donc de copier-coller-renommer, en une seule ligne de commande !

Avec l'aide de l'option <code>-r</code> vous pouvez copier un dossier et tout son contenu au lieu d'un simple fichier.

Essayons ! Assurez vous de savoir où vous êtes. Puis copier le fichier <code>/shared/bank/homo_sapiens/hg38/fasta/hg38.fa.fai</code> dans le sous-dossier <code>ref</code>.

```bash
# Savoir où l'on est
pwd
# Copier le fichier uniquement dans le dossier rna-seq/ref
cp /shared/bank/homo_sapiens/hg38/fasta/hg38.fa.fai rna-seq/ref/
```

Affichons un arbre de notre dossier rna-seq pour nous assurer que tout est bon.

```bash
# Afficher l'arbre du dossier rna-seq uniquement
tree rna-seq
```

### Couper/coller un fichier

<code>mv</code> est l'équivalent de couper/coller mais aussi de renommer.

Cette commande a donc deux usages : 

- Renommer :
<code>mv fichier_à_renommer nouveau_nom</code>

- Couper/coller :
<code>mv fichier_source destination</code>

<code>mv</code> est une commande puissante, elle peut renommer un dossier sans option, faites donc bien attention.

Testons : il faut que vous déplaciez le fichier <code>hg38.fa.fai</code> dans le dossier <code>output</code>.

```bash
# Déplacer le fichier
mv rna-seq/ref/hg38.fa.fai rna-seq/output/
```

Un petit arbre pour vérifier

```bash
# Arbre du dossier rna-seq
tree rna-seq
```

Et si depuis le début nous faisions fausse route ? Le dossier <code>rna-seq</code> devait bien s'appeler <code>chip-seq</code> ! Renommez-le.

```bash
# Renommer le dossier rna-seq en chip-seq
mv rna-seq chip-seq
```

Listons le contenu du dossier où nous sommes pour vérifier que tout est bon.

```bash
# Lister le contenu
ls -r
```

### Supprimer un fichier ou dossier non vide

C'est sans nul doute la commande la plus traîte qui a fait pleurer au moins une fois chacun d'entre nous.

<code>rm</code> permet de supprimer un fichier et accompagnée de son option <code>-r</code> elle supprimera même un dossier.

Nous avons donc :

<code>rm fichier</code> pour la suppression d'un fichier

<code>rm -r dossier</code> pour la suppression (suspens...) d'un dossier !

Elle ne vous demandera pas avant suppression si vous êtes VRAIIIIMENT sûrs de vous. Mieux vaux donc relire deux fois le nom du fichier que l'on est sur le point de supprimer !

##### Tips : Vous pouvez faire un <code>ls</code> du fichier que vous souhaitez supprimer afin de vous assurer que c'est le bon. Ensuite vous modifiez simplement le <code>ls</code> en <code>rm</code> 

Maintenant que vous êtes mis en garde, supprimons ce fichier <code>hg38.fa.fai</code> qui ne nous sert finalement pas...

```bash
# Supprimer le fichier
rm chip-seq/output/hg38.fa.fai
```

Vérifions une dernière fois que tout est bon (avec un arbre par exemple ?)

```bash
# Tracer l'arbre du dossier chip-seq
tree chip-seq
```

## Histoire de droits (Bonus)

Voici quelques informations complémentaires sur le système de droit sous Unix :

Il se peut qu'un message d'erreur apparaisse vous annonçant que vous n'avez pas les droits.

Ces droits sont régis pas 3 lettres : <code>r</code>, <code>w</code> et <code>x</code>.

- <code>r</code> : **r**ead, droit de lire un fichier, de l'ouvrir.
- <code>w</code> : **w**rite, droit d'écriture et de modification d'un fichier.
- <code>x</code> : e**x**ecute, droit d'exécuter un fichier. C'est surtout valable pour les scripts. 

Lorsque vous faites un <code>ls -l</code>, vous pouvez noter que certaines lignes commencent par un <code>d</code>, signifiant dossier.

Notons également le fait que les rwx se répètent, ou non, trois fois. 

- Le premier triplet correspond aux droits que possèdent le propriétaire du fichier, 
- Le second correspond aux droits alloués aux utilisateurs du même groupe que le propriétaire du fichier, 
- Le dernier correspond aux droits que possèdent tous les autres utilisateurs.
